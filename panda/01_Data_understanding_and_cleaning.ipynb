###Data Columns Description:
* List Sr.No. - Serial number of the player in the auction list.
* Set No. - Grouping of players into sets for auction purposes.
* 2025 Set - Specific set identifier for 2025 auction.
* First Name - Player's first name.
* Surname - Player's last name.
* Country - Nationality of the player.
* State Association - State or regional cricket association.
* DOB - Date of Birth of the player.
* Age - Player’s age as of 2025.
* Specialism - Player’s role (e.g., Batter, Bowler, All-Rounder, Wicketkeeper).
* Test Caps - Number of Test matches played by the player.
* ODI Caps - Number of One Day International matches played.
* T20 Caps - Number of T20 International matches played.
* IPL - Number of IPL matches played by the player.
* Previous IPL Team(s) - List of IPL teams the player has been associated with.
* 2024 Team - Player's team during the 2024 IPL season.
* 2024 IPL - Performance metrics or appearances in the 2024 season.
* C/U/A - Capped/Uncapped/Associate player status.
* Reserve Price Rs Lakh - Base price for the auction, in Indian Rupees (Lakh).

```python
#data cleaning
df['State Association'] = df['State Association'].fillna('International')
df['Bowling Skills'] = df['Bowling Skills'].fillna('Can not bowl ')
df['Test caps'] = df['Test caps'].fillna(0)
df['ODI caps'] = df['ODI caps'].fillna(0)
df['T20 caps'] = df['T20 caps'].fillna(0)
df['IPL caps'] = df['IPL caps'].fillna(0)
df['Previous IPLTeam(s)'] = df['Previous IPLTeam(s)'].fillna('first time in IPL')
df['2024 Team'] = df['2024 Team'].fillna('do not play in 2024')
df['2024 IPL'] = df['2024 IPL'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['full name'] = df['First Name'] + ' ' + df['Surname']
```

```python
col = df.pop('full name')
df.insert(df.columns.get_loc('Surname') + 1, 'full name', col)
```

```python
caps_cols = ['Test caps', 'ODI caps', 'T20 caps', 'IPL caps']
df[caps_cols] = df[caps_cols].astype('int64')
df['DOB'] = pd.to_datetime(df['DOB'], format='%d-%m-%Y')
```

```python
df.to_excel('TATA-IPL-2025-Auction-List-cleaned.xlsx', index=False)
```
